```python
import numpy as np
import pandas as pd
from os import listdir
from os.path import isfile, join
```

```python
#Get files
mypath = 'C:\\Users\\janol\\OneDrive\\Documentos\\DragApp\\KW Landscape English Language\\'

onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
#print(onlyfiles)

large_df = pd.DataFrame()                    

for x in onlyfiles:
    df = pd.read_csv(mypath + x, sep='\r\t\n\0', engine='python', skiprows=(0,1), header=None)
    
    #Fix import #1 separate columns
    series = df.squeeze()
    series = series.str.replace('\x00', '', regex=True)
    df = series.str.split("\t", expand=True,)

    #Fix import #2 column names
    new_header = df.iloc[0]
    df = df[1:]
    df.columns = new_header
    df = df.iloc[: , :24]
    #print(df.shape)
    large_df = large_df.append(df)


print(large_df.shape)
#print(large_df.columns)
display(large_df.head())
```

```python
print(type(large_df))

large_df.to_csv('C:\\Users\\janol\\OneDrive\\Documentos\\DragApp\\KW_Analysis_EN.csv', encoding='utf-8-sig')
```
